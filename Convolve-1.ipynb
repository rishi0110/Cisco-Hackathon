```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import nltk
```

```python
vocab_size = 10000
embedding_dim = 16
max_length = 10
trunc_type = 'post'
padding_type = 'post'
oov_tok = '<OOV>'
```

```python
tokenizer = Tokenizer(num_words = vocab_size,oov_token = oov_tok)
tokenizer.fit_on_texts(X_train)
word_index = tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(X_train)
X_train_final = pad_sequences(sequences,maxlen = max_length,truncating = trunc_type)
testing_sentences = tokenizer.texts_to_sequences(X_test)
X_test_final = pad_sequences(testing_sentences,maxlen = max_length,truncating = trunc_type)
```
